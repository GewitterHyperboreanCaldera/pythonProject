```python
import numpy as np
import scipy.linalg as la

A = np.array([[8, 3, -2], [-4, 7, 5], [3, 4, -12]])
b = np.array([9, 15, 35])

P, L, U = la.lu(A)

# First solve Ly = b
y = la.solve_triangular(L, b, lower=True)

# Then solve Ux = y
x = la.solve_triangular(U, y)

print(x)
```

```python
import numpy as np
import seaborn as sns
from scipy.linalg import lu
from matplotlib import pyplot as plt

A=np.array([[5,2,-2,3],
            [-2,5,-8,7],
            [7,-5,1,-6],
            [-5,4,-4,8]])

P,L,U= lu(A, permute_l=False)#lu decomposition with permutation matrix
```

```python
import matplotlib.pyplot as plt  # 导入Matplotlib库，用于绘图
import numpy as np  # 导入NumPy库，用于数值计算

p_values = [0.05, 0.2, 0.5, 1, 1.5, 2, 4, 8, np.inf]  # 定义不同的p值

x1 = np.linspace(-2.5, 2.5, num=101)  # 定义x1的范围
x2 = x1  # 设置x2的范围与x1相同

xx1, xx2 = np.meshgrid(x1, x2)  # 生成x1和x2的网格

fig, axes = plt.subplots(ncols=3, nrows=3, figsize=(12, 12))  # 创建3x3子图

## 遍历p值并绘制Lp范数的等高线图
for p, ax in zip(p_values, axes.flat):  # 遍历每个p值和子图轴

    if np.isinf(p):  # 检查p是否为无穷大
        zz = np.maximum(np.abs(xx1), np.abs(xx2))  # L∞范数计算
    else:
        zz = ((np.abs((xx1))**p) + (np.abs((xx2))**p))**(1./p)  # Lp范数计算

    ## 绘制Lp的等高线图
    ax.contourf(xx1, xx2, zz, 20, cmap='RdYlBu_r')  # 使用填充的等高线图展示Lp范数

    ## 绘制Lp=1的等高线
    ax.contour(xx1, xx2, zz, [1], colors='k', linewidths=2)  # 绘制Lp=1的等高线，颜色为黑色

    ## 图形装饰
    ax.axhline(y=0, color='k', linewidth=0.25)  # 绘制y=0的水平线
    ax.axvline(x=0, color='k', linewidth=0.25)  # 绘制x=0的垂直线
    ax.set_xlim(-2.5, 2.5)  # 设置x轴范围
    ax.set_ylim(-2.5, 2.5)  # 设置y轴范围
    ax.spines['top'].set_visible(False)  # 隐藏顶部边框
    ax.spines['right'].set_visible(False)  # 隐藏右侧边框
    ax.spines['bottom'].set_visible(False)  # 隐藏底部边框
    ax.spines['left'].set_visible(False)  # 隐藏左侧边框
    ax.set_xlabel('$x_1$')  # 设置x轴标签
    ax.set_ylabel('$x_2$')  # 设置y轴标签
    ax.set_title('p = ' + str(p))  # 设置子图标题，显示当前的p值
    ax.set_aspect('equal', adjustable='box')  # 设置坐标轴比例为相等

plt.tight_layout()
plt.show()  # 显示图形
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#1.generating virtual dataset
X, y = make_classification(
    n_samples=3000,
    n_features=8,
    n_informative=5,
    n_redundant=2,
    n_classes=3,
    n_clusters_per_class=1,
    class_sep=2,
    random_state=42
)

#2.data standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#3.PCA
pca = PCA(n_components=8)
X_pca = pca.fit_transform(X_scaled)

#4.construct DataFrame to visualize
pca_df = pd.DataFrame(
    data = X_pca,
    columns= [f"PC{i+1}" for i in range(X_pca.shape[1])]
)
pca_df["target"] = y

#ploting style
sns.set(style="whitegrid", palette="bright", font_scale=1.1)
plt.rcParams['figure.figsize'] = (8,5)

#explaining variance ratio
plt.figure()
sns.barplot(
    x= [f"PC{i+1}" for i in range (len(pca.explained_variance_ratio_))],
    y = pca.explained_variance_ratio_,
    palette = "Set2"
)

plt.title("Each PCA's explained variance ratio")
plt.ylabel("variance explaining rate")
plt.xlabel("PCA")
plt.tight_layout()
plt.show()

#scatter plot between PC1 and PC2
plt.figure()
sns.scatterplot(
    data = pca_df,
    x    = "PC1", y    ="PC2",
    hue = "target",
    palette= "Set1"
)
plt.title("PCA projection:PC1 vs PC2")
plt.ylabel("PCA 1")
plt.xlabel("PCA 2")
plt.legend(title ="type")
plt.tight_layout()
plt.show()

#scatter plot between PC3 and PC4
plt.figure()
sns.scatterplot(
    data = pca_df,
    x    = "PC3", y    ="PC4",
    hue = "target",
    palette= "Set1"
)
plt.title("PCA projection:PC3 vs PC4")
plt.ylabel("PCA 3")
plt.xlabel("PCA 4")
plt.legend(title ="type")
plt.tight_layout()
plt.show()

# explaining total variance curve
plt.figure()
plt.plot(
    np.cumsum(pca.explained_variance_ratio_),
    marker = 'o'
)
plt.title("total PCA explained ratio")
plt.xlabel("total PCA")
plt.ylabel("total explained rate")
plt.grid(True)
plt.tight_layout()
plt.show()

components_df = pd.DataFrame(
    pca.components_,
    columns = [f"Feature{i+1}" for i in range(X.shape[1])],
    index   = [f"PC{i+1}" for i in range(X.shape[1])]
)

plt.figure(figsize=(10,6))
sns.heatmap(
    components_df,
    cmap   = "Spectral",
    annot  = True,
    fmt    = ".2f"
)
plt.title("relationship between PCA and original characteristics(weight)")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

# Define a simple vortex field (cyclonic flow)
# Velocity components: u = -y / r, v = x / r, where r = sqrt(x^2 + y^2) + epsilon to avoid division by zero
epsilon = 0.1
r = np.sqrt(X**2 + Y**2) + epsilon
u = -Y / r  # Eastward component
v = X / r   # Northward component

# Null point simulation: add a perturbation to create a saddle-like null point near (1,0)
u += 0.5 * (X - 1)  # Add a divergent field
v += -0.5 * Y

# Plot the vector field
fig, ax = plt.subplots(figsize=(8, 6))
ax.quiver(X, Y, u, v, color='b')
ax.set_title('Simplified Vector Field Simulating Cyclone with Null Point')
ax.set_xlabel('X (East-West)')
ax.set_ylabel('Y (North-South)')
ax.grid(True)

# Save the plot to a file (since we can't display it directly)
plt.savefig('vector_field.png')
print("Vector field plot saved as 'vector_field.png'")
```
